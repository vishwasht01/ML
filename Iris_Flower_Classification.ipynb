Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score,classification_report
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

Loading the dataset

```python
iris_data = pd.read_csv('/content/iris_data.csv',header=None)
x=iris_data.drop([4],axis=1)
y=iris_data[4]
```

```python
print(x)
print(y)
```

Splitting dataset

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,stratify=y,random_state=1)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

Removing the mean and scaling to unit variance using StandardScaler

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
print(x_test_scaled)
```

Applying PCA for dimensionality reduction

```python
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train_scaled)
x_test_pca = pca.fit_transform(x_test_scaled)
```

```python
#This is an attribute of the PCA object in scikit-learn. It's a NumPy array where each element represents the percentage of the variance in the original data that is explained by each corresponding principal component
print("Explained variance ratio by PCA components:", pca.explained_variance_ratio_)
```

```python
svm_model=SVC(kernel='linear',C=1.0,gamma=0.1,random_state=42)
svm_model.fit(x_train_pca,y_train)
y_pred_svm=svm_model.predict(x_test_pca)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train_pca,y_train)
y_pred_knn = knn_model.predict(x_test_pca)
```

```python
acc_svm=accuracy_score(y_test,y_pred_svm)
acc_knn=accuracy_score(y_test,y_pred_knn)
```

```python

print("\nComparison:")
print(f"SVM Accuracy: {acc_svm*100:.2f}%")
print(f"KNN Accuracy: {acc_knn*100:.2f}%")
```
